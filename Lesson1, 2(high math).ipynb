#### Урок 1, задание 4 (в программе): 
Постройте на одном графике две кривые y(x) для функции двух переменной
y(k,x)=cos(k·x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 121)
def y(k, x):
    return np.cos(k * x)
```

```python
plt.plot(x, y(1, x), label='k=1', color='brown')
plt.plot(x, y(4, x), label='k=4', color='green')
plt.xlabel('x')
plt.ylabel('y(k,x)=cos(k*x)')
plt.legend()
```

#### Урок 2, задание 1 (в программе):
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10) 1) Найдите их сумму. (на листочке) 2) 
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе) 

```python
a = np.array([10, 10, 10])
b = np.array([0, 0, -10])
```

```python
def vector_length(vector):
    return np.sqrt(np.sum(np.power(vector, 2)))
```

```python
print(f'Длина вектора {a}: {vector_length(a)}')
print(f'Длина вектора {b}: {vector_length(b)}')
```

#### 3. Задание (в программе) Напишите код на Python, реализующий построение графиков: 1. окружности, 2. эллипса, 3. гиперболы.

```python
# circle
t = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(t), np.sin(t))

plt.axes().set_aspect('equal')
plt.show()
```

```python
# ellipse
r1 = 3
r2 = 2

t = np.linspace(0, 2*np.pi, 100)
plt.plot(r1*np.cos(t), r2*np.sin(t))

plt.axes().set_aspect('equal')
plt.show()
```

```python
# hyperbola
X = np.around(np.arange(-10, 10, 0.1), decimals=1)
Y = []
for x in X:
    if x != 0:
        Y.append(1/x)
    else:
        Y.append(np.nan)
 
plt.plot(X, Y)
plt.show()
```

#### 5. Задание (в программе) 
1) Нарисуйте трехмерный график двух параллельных плоскостей.2) Нарисуйте трехмерный график двух любых поверхностей второго порядка. 

```python
from mpl_toolkits.mplot3d import Axes3D
from pylab import *
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

Z = 3*X + 3*Y
ax.plot_surface(X, Y, Z)

D = 30
Z2 = 3*X + 3*Y + D 
ax.plot_surface(X, Y, Z2)

plt.show()
```

```python
fig = plt.figure(figsize=plt.figaspect(0.4))

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

#Конус:
ax = fig.add_subplot(1, 2, 1, projection='3d')
Z = np.sqrt(X**2 + Y**2)
ax.plot_surface(X, Y, Z)
ax.set_title('Конус')

#Элиптический парабалойд:
ax = fig.add_subplot(1, 2, 2, projection='3d')
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z)
ax.set_title('Элиптический парабалойд')

plt.show()
```
